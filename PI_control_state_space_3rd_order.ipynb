```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot time response of a system in state space representation
# state space model
# xdot= A * x + B * u
# y   = C * x + D * u

A = [[   0,  1,   0],
     [   0,  0,   1],
     [ -24, -26, -9]]

B = [[0],
     [0],
     [1]]
C = [[2, 7, 1]]
D = [[0]]
```

```python
dt = 0.001 # simulation step size
t = np.arange(0,15,dt)
x = [[0],
     [0],
     [0]]
r = 1                  # reference iput
kp = 0.05
ki = 0.05
u = 0.0
#print(x.reshape(-1,1))
print(t.shape[0])
rplot = np.zeros(t.shape[0])
yplot = np.zeros(t.shape[0])
sum_e = 0

for i in range(t.shape[0]):
  if i >= 5000:
    r = 2
  if i >= 10000:
    r = 0.5
  #print(i)
  xdot=np.dot(A, x)+np.dot(B,u)
  #print(xdot)
  #print(xdot.shape)
  x = x + xdot*dt
  y = np.dot(C,x)+np.dot(D,u)
  e = r - y
  u = kp * e + ki * sum_e
  sum_e = sum_e + e
  yplot[i] = y
  rplot[i] = r

plt.plot(t,rplot)
plt.plot(t,yplot)
plt.grid()
plt.legend(['Ref','Output'])
plt.show()

```
